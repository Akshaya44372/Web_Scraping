```python
page=requests.get(Url1)
```

```python
page
```

```python
#removes the unwanted text
soup=BeautifulSoup(page.text)
```

```python
brand=[]
price_data=[]
rating_data=[]
pages=40
```

```python
# re means regular expression
import re
import numpy as np
```

```python
#to remove repeated values and also doesn't prints the symbols and prints the brand name 
brand=[]
for i in soup.find_all("a",class_="pIpigb"):
    print(i.text)
    regex=re.findall("^[a-zA-Z]+",i.text)
    print(regex)
    if regex:
        brand.append(regex[0])
    else:
        brand.append(np.nan)
```

```python
rating = []

for i in soup.find_all("div", class_="MKiFS6"):
    print(i.text)
    rating.append(i.text)
```

```python
selling_price=[]
for i in soup("div",class_="hZ3P6w"):
    print(i.text)
    selling_price.append(i.text)
```

```python
original_price=[]
for i in soup("div",class_="kRYCnD"):
    print(i.text)
    original_price.append(i.text)
```

```python
stock = []
for i in soup.find_all("div", class_="HZ0E6r Rm9_cy"):
    print(i.text)
    stock.append(i.text)
```

```python
ratings_reviews_count = [] 
for i in soup.find_all("span", class_="PvbNMB"): 
    print(i.text) 
    ratings_reviews_count.append(i.text)
```

```python
discounts = []

for i in soup.find_all("div", class_="HQe8jr"):
    print(i.text)
    discounts.append(i.text)
```

```python
len(brand)
```

```python
len(rating)
```

```python
len(selling_price)
```

```python
len(original_price)
```

```python
len(stock)
```

```python
len(ratings_reviews_count)
```

```python
len(discounts)
```

```python
brand = brand[:27]
rating = rating[:27]
selling_price = selling_price[:27]
original_price = original_price[:27]
stock = stock[:27]
discounts = discounts[:27]
```
